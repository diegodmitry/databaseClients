```python
from mysql.connector.pooling import MySQLConnectionPool

dbconfig = {
    "database": "little_lemon_db",
    "user": "root",
    "password": "mysecret"
}

# Tasks

# Task 1
try:
    pool_a = MySQLConnectionPool(pool_name="pool_a",
                                pool_size=2,
                                **dbconfig)
    print("Pool created successfully.")
except connector.Error as e:
    print(f"Error while creating connection pool: {e}")
finally:
    cursor.close()
    connection.close()
```

```python
# Task 2
connection = pool_a.get_connection()
cursor = connection.cursor()

try:
    # Splitting the DROP and CREATE statements
    cursor.execute("DROP PROCEDURE IF EXISTS PeakHours")
    create_procedure_query = """
    CREATE PROCEDURE PeakHours()
    BEGIN
        SELECT HOUR(BookingSlot) AS BookingHour, COUNT(*) AS NumberOfBookings
        FROM Bookings
        GROUP BY BookingHour
        ORDER BY NumberOfBookings DESC;
    END
    """
    cursor.execute(create_procedure_query)
    connection.commit()
    print("Stored procedure 'PeakHours' created successfully.")
except Error as e:
    print(f"Error creating stored procedure: {e}")
finally:
    # Close cursor and connection
    cursor.close()
    connection.close()

# To call and fetch results from the stored procedure, ensure to open a new connection and cursor from the pool
connection = pool_a.get_connection()
cursor = connection.cursor()

try:
    # Calling the stored procedure
    cursor.callproc('PeakHours')
    print("Stored procedure 'PeakHours' called successfully.")

    result_sets = cursor.stored_results()
    for result in result_sets:
        # fetch the column names
        column_names = result.description
        if column_names is not None:
            column_headers = [i[0] for i in column_names]
            print(" | ".join(column_headers))

        # Fetching the results
        dataset = result.fetchall()
    
        # Print the sorted data
        for row in dataset:
            print(" | ".join(str(value) for value in row))
except connector.Error as e:
    print(f"Error calling stored procedure: {e}")
finally:
    # Always ensure resources are released
    cursor.close()
    connection.close()
```

```python
import mysql.connector as connector

connection=connector.connect(
    host="localhost",
    port=3306,
    user="root",
    password="mysecret")

connection = connector.connect(**dbconfig)
cursor = connection.cursor()

# Task 3
try:
    cursor.execute("DROP PROCEDURE IF EXISTS GuestStatus;")
    connection.commit()

    create_procedure_gueststatus = """
    CREATE PROCEDURE GuestStatus()
    BEGIN
        SELECT CONCAT(GuestFirstName, ' ', GuestLastName) AS GuestName,
            CASE
                WHEN Role IN ('Manager', 'Assistant Manager') THEN 'Ready to pay'
                WHEN Role = 'Head Chef' THEN 'Ready to serve'
                WHEN Role = 'Assistant Chef' THEN 'Preparing Order'
                WHEN Role = 'Head Waiter' THEN 'Order served'
                ELSE 'Unknown status'
            END AS OrderStatus
        FROM Bookings
        LEFT JOIN Employees ON Bookings.EmployeeID = Employees.EmployeeID
        ORDER BY GuestName ASC;
    END;
    """
    cursor.execute(create_procedure_gueststatus)
    connection.commit()
    print("Stored procedure 'GuestStatus' created successfully.")
except connector.Error as e:
    print(f"Error creating stored procedure: {e}")
    
cursor.close()

try:
    cursor = connection.cursor()  # Open a new cursor for calling the stored procedure
    cursor.callproc('GuestStatus')
    print("Stored procedure 'GuestStatus' called successfully.")

    # Fetch the results
    for result in cursor.stored_results():
        dataset = result.fetchall()
        column_names = [desc[0] for desc in result.description]  # Extract column names

        # Print column names
        print(" | ".join(column_names))

        # Print data rows
        print("dataset:", dataset)
        for row in dataset:
            print(" | ".join(str(value) for value in row))
except connector.Error as e:
    print(f"Error calling/fetching stored procedure: {e}")
finally:
    cursor.close()
    connection.close()
    print("Cursor and connection closed.")
```
