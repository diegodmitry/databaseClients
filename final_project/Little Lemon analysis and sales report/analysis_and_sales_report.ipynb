```python
from mysql.connector.pooling import MySQLConnectionPool
from mysql.connector import Error, PoolError

dbconfig = {
    "database": "little_lemon_db",
    "user": "root",
    "password": "mysecret",
    "host": "localhost"
}

try:
    pool_b = MySQLConnectionPool(pool_name="pool_b",
                                    pool_size=2,
                                    **dbconfig)
    print("Connection pool 'pool_b' with 2 connections was created successfully.")
except Error as e:
    print(f"Error while creating connection pool: {e}")

bookings = [
    (8, "Anees", "Java", "18:00", 6),
    (5, "Bald", "Vin", "19:00", 6),
    (12, "Jay", "Kon", "19:30", 6)
]

for booking in bookings:
    try:
        connection = pool_b.get_connection()
    except PoolError as e:
        print(f"Could not obtain a connection for {booking[1]}: {e}")
        continue  # Skip this booking

    try:
        cursor = connection.cursor()
        insert_query = """
            INSERT INTO Bookings (TableNo, GuestFirstName, GuestLastName, BookingSlot, EmployeeID)
            VALUES (%s, %s, %s, %s, %s);
        """
        cursor.execute(insert_query, booking)
        connection.commit()
        print(f"Booking for {booking[1]} successfully inserted.")
    except Error as e:
        print(f"Error inserting booking for {booking[1]}: {e}")
    # finally:
    #     cursor.close()
    #     connection.close()
```

```python
connection = pool_b.get_connection()
cursor = connection.cursor()

try:
    # Name and EmployeeID
    cursor.execute("SELECT Name, EmployeeID FROM Employees WHERE Role = 'Manager';")
    for (name, employee_id) in cursor:
        print(f"Manager: {name}, EmployeeID: {employee_id}")

    # Name and role
    cursor.execute("SELECT Name, Role FROM Employees ORDER BY Annual_Salary DESC LIMIT 1;")
    for (name, role) in cursor:
        print(f"Highest Paid Employee: {name}, Role: {role}")

    # Number of guests booked
    cursor.execute("SELECT COUNT(*) AS NumberOfGuests FROM Bookings WHERE BookingSlot BETWEEN '18:00' AND '20:00';")
    for (number_of_guests,) in cursor:
        print(f"Number of Guests Booked Between 18:00 and 20:00: {number_of_guests}")

    # Full name and BookingID of all guests waiting
    cursor.execute("SELECT CONCAT(GuestFirstName, ' ', GuestLastName) AS GuestName, BookingID FROM Bookings ORDER BY BookingSlot;")
    print("Guests Waiting to be Seated:")
    for (guest_name, booking_id) in cursor:
        print(f"Guest: {guest_name}, BookingID: {booking_id}")

except PoolError as e:
    print(f"Could not obtain a connection: {e}")
except Error as e:
    print(f"An error occurred: {e}")
finally:
    cursor.close()
    connection.close()
```

```python
try:
    connection = pool_b.get_connection()
    cursor = connection.cursor()
    
    cursor.execute("DROP PROCEDURE IF EXISTS BasicSalesReport;")
    
    create_procedure_query = """
    CREATE PROCEDURE BasicSalesReport()
    BEGIN
        SELECT 
            SUM(BillAmount) AS TotalSales,
            AVG(BillAmount) AS AverageSale,
            MIN(BillAmount) AS MinimumBill,
            MAX(BillAmount) AS MaximumBill
        FROM Orders;
    END;
    """
    
    print("Stored procedure 'BasicSalesReport' created.")
    cursor.execute(create_procedure_query)
    cursor.callproc('BasicSalesReport')
    for result in cursor.stored_results():
        stats = result.fetchall()
        print("Total Sales | Average Sale | Minimum Bill | Maximum Bill")
        for stat in stats:
            print(f"{stat[0]} | {stat[1]} | {stat[2]} | {stat[3]}")
            
    connection.commit()

except Error as e:
    print(f"Error creating stored procedure: {e}")
finally:
    cursor.close()
    connection.close()
```

```python
try:
    connection = pool_b.get_connection()
    cursor = connection.cursor(buffered=True)
    
    query = """
    SELECT 
        B.BookingSlot, 
        CONCAT(B.GuestFirstName, ' ', B.GuestLastName) AS GuestName,
        CONCAT(E.Name, ' [', E.Role, ']') AS EmployeeInfo
    FROM 
        Bookings B
    JOIN 
        Employees E ON B.EmployeeID = E.EmployeeID
    ORDER BY 
        B.BookingSlot ASC
    LIMIT 3;
    """
    
    cursor.execute(query)
    
    for (booking_slot, guest_name, employee_info) in cursor:
        print(f"[{booking_slot}]")
        print(f"[{guest_name}]")
        print(f"[Assigned to: {employee_info}]")
        print("")  # Adding a newline for readability between bookings
        
except Error as e:
    print(f"Error retrieving upcoming bookings: {e}")
finally:
    cursor.close()
    connection.close()
```
